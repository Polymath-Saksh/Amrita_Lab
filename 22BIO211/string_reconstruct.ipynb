```python
def generate_kmer_composition(string, k):
    kmers = []
    for i in range(len(string) - k + 1):
        kmer = string[i:i+k]
        kmers.append(kmer)

    kmers = sorted(kmers)
    return kmers

if __name__=='__main__':
    x = input('Enter the string: ')
    k = int(input('Enter the value of k: '))
    print(f'Input string: {x}')
    print('Kmer composition:')
    for kmer in generate_kmer_composition(x, k):
        print(kmer)
```

```python
# String Reconstruction Problem: Reconstruct a string from its k-mer composition.
# Construct the De Bruijn Graph of a String Problem: Construct the de Bruijn graph of a string.
# Input: An integer k and a collection Patterns of k-mers.
def de_bruijn_graph(patterns, k):
    de_bruijn = {}
    for pattern in patterns:
        prefix = pattern[:-1]
        suffix = pattern[1:]
        if prefix in de_bruijn:
            de_bruijn[prefix].append(suffix)
        else:
            de_bruijn[prefix] = [suffix]
    return de_bruijn

if __name__=='__main__':
    k = int(input('Enter the value of k: '))
    ls = input('Enter the kmer composition: ').split()
    print('De Bruijn Graph:')
    for key, value in de_bruijn_graph(ls, k).items():
        print(f'{key} -> {",".join(value)}')
```
