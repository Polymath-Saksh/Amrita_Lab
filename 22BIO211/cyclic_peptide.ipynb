````python

# Implement a Python function to calculate the score of a cyclic peptide against a given experimental spectrum.

def calculate_score(peptide, spectrum):
    score = 0
    theoretical_spectrum = generate_theoretical_spectrum(peptide)

    for mass in theoretical_spectrum:
        if mass in spectrum:
            score += 1
            spectrum.remove(mass)

    return score

def generate_theoretical_spectrum(peptide):
    mass_table = {
        'G': 57, 'A': 71, 'S': 87, 'P': 97, 'V': 99,
        'T': 101, 'C': 103, 'I': 113, 'L': 113, 'N': 114,
        'D': 115, 'K': 128, 'Q': 128, 'E': 129, 'M': 131,
        'H': 137, 'F': 147, 'R': 156, 'Y': 163, 'W': 186
    }

    prefix_mass = [0]
    for i in range(1, len(peptide) + 1):
        prefix_mass.append(prefix_mass[i - 1] + mass_table[peptide[i - 1]])

    peptide_mass = prefix_mass[-1]
    theoretical_spectrum = [0]

    for i in range(len(peptide)):
        for j in range(i + 1, len(peptide) + 1):
            theoretical_spectrum.append(prefix_mass[j] - prefix_mass[i])

            if i > 0 and j < len(peptide):
                theoretical_spectrum.append(peptide_mass - (prefix_mass[j] - prefix_mass[i]))

    return sorted(theoretical_spectrum)

if __name__ == '__main__':
    peptide = 'NQEL'
    spectrum = [0, 99, 113, 114, 128, 227, 257, 299, 355, 356, 370, 371, 484]
    print(calculate_score(peptide, spectrum))
````

````python
# Function to convert DNA sequence to peptide
def dna_to_peptide(dna):
    # Define the codon table
    codon_table = {
        'TTT': 'F', 'TTC': 'F', 'TTA': 'L', 'TTG': 'L',
        'TCT': 'S', 'TCC': 'S', 'TCA': 'S', 'TCG': 'S',
        'TAT': 'Y', 'TAC': 'Y', 'TAA': '', 'TAG': '',
        'TGT': 'C', 'TGC': 'C', 'TGA': '', 'TGG': 'W',
        'CTT': 'L', 'CTC': 'L', 'CTA': 'L', 'CTG': 'L',
        'CCT': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
        'CAT': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
        'CGT': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
        'ATT': 'I', 'ATC': 'I', 'ATA': 'I', 'ATG': 'M',
        'ACT': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
        'AAT': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K',
        'AGT': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R',
        'GTT': 'V', 'GTC': 'V', 'GTA': 'V', 'GTG': 'V',
        'GCT': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
        'GAT': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',
        'GGT': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'
    }
# Initialize the peptide string
    peptide = ''
    # Iterate over the DNA sequence, three bases at a time
    for i in range(0, len(dna), 3):
        codon = dna[i:i + 3]
        if codon in codon_table:
            aa = codon_table[codon]
            # If the amino acid is a stop codon, break the loop
            if aa == '':
                break
            # Otherwise, add the amino acid to the peptide
            peptide += aa
        else:
            # If the codon is not in the table, break the loop
            break
    return peptide

# Function to find all substrings of a DNA sequence that encode a given peptide
def find_substrings(text, peptide):
    # Initialize the list of substrings
    substrings = []

    # Get the lengths of the peptide and the DNA sequence
    peptide_len = len(peptide)
    dna_len = len(text)

    # Iterate over the DNA sequence
    for i in range(dna_len - 3 * peptide_len + 1):
        sub_seq = text[i:i + 3 * peptide_len]
        translated_peptide = dna_to_peptide(sub_seq)
        # If the translated peptide is the same as the given peptide
        if translated_peptide == peptide:
            substrings.append(sub_seq)

    # Return the list of substrings
    return substrings

# Main function
if __name__ == "__main__":
    # Define the DNA sequence and the peptide
    text = "ATGGCCATGGCCCCCAGAACTGAGATCAATAGTACCCGTATTAACGGGTGA"
    peptide = "MA"

    # Find all substrings of the DNA sequence that encode the peptide
    substrings = find_substrings(text, peptide)
    if substrings:
        print("Unique Substrings encoding the peptide:")
        # Remove duplicates from the list of substrings and print them
        substrings = list(set(substrings))
        for substring in substrings:
            print(substring)
    else:
        # If there are no substrings, print a message
        print("No substrings found encoding the peptide.")
````
