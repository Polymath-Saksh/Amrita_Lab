## Generating Theoretical Spectrum of a Cyclic Peptide

```python
# Generate all subpeptides

def subpeptides(peptide):
    length = len(peptide)
    subpeptides = []
    for i in range(length):
        for j in range(1, length):
            subpeptides.append(peptide[i:i+j])

    subpeptides.append(peptide)
    
    return subpeptides
```

```python
# Calculate the mass of each subpeptide based on the given mass of each amino acid. Use the mass of each amino acid to determine the mass of the subpeptides.
def mass(peptide):
    mass_table = {
        'G': 57,
        'A': 71,
        'S': 87,
        'P': 97,
        'V': 99,
        'T': 101,
        'C': 103,
        'I': 113,
        'L': 113,
        'N': 114,
        'D': 115,
        'K': 128,
        'Q': 128,
        'E': 129,
        'M': 131,
        'H': 137,
        'F': 147,
        'R': 156,
        'Y': 163,
        'W': 186
    }

    mass = 0
    for amino_acid in peptide:
        mass += mass_table[amino_acid]

    return mass
# Combine the masses of all subpeptides to form the theoretical spectrum. Include the masses of all subpeptides.

def theoretical_spectrum(peptide):
    subpeptides_list = subpeptides(peptide)
    spectrum = []
    for subpeptide in subpeptides_list:
        spectrum.append(mass(subpeptide))

    return sorted(spectrum)

if __name__ == '__main__':
    peptide = 'NRL'
    print(' '.join(map(str, theoretical_spectrum(peptide))))
```
