```python
mass_table = {
    'G': 57, 'A': 71, 'S': 87, 'P': 97, 'V': 99,
    'T': 101, 'C': 103, 'I': 113, 'L': 113, 'N': 114,
    'D': 115, 'K': 128, 'Q': 128, 'E': 129, 'M': 131,
    'H': 137, 'F': 147, 'R': 156, 'Y': 163, 'W': 186
}

def subpeptides(peptide):
    subpeptides = [peptide]
    peptide_length = len(peptide)
    for i in range(peptide_length): # i is the starting index of the subpeptide
        for j in range(1, peptide_length): # j is the length of the subpeptide
            if i + j <= peptide_length: # if the subpeptide is within the peptide
                subpeptides.append(peptide[i:i+j])
            else:# if the subpeptide is not within the peptide
                subpeptides.append(peptide[i:] + peptide[:i + j - peptide_length])
    
    return subpeptides
```

```python
def theoretical_spectrum(peptide):
    spectrum = [0]
    peptide_length = len(peptide) # length of the peptide
    for i in range(peptide_length): # for each amino acid in the peptide
        for j in range(1, peptide_length): # for each subpeptide length
            if i + j <= peptide_length: 
                subpeptide = peptide[i:i+j]
            else:
                subpeptide = peptide[i:] + peptide[:i + j - peptide_length]
            spectrum.append(sum([mass_table[aa] for aa in subpeptide]))
    return sorted(spectrum)
```

```python
reverse_mass_table = {v: k for k, v in mass_table.items()}

def compare_spectrum(theoretical_spectrum, experimental_spectrum):
    matching_peptides = []
    for mass in theoretical_spectrum: # Iterate over the theoretical spectrum
        if mass in experimental_spectrum: # Check if the mass is in the experimental spectrum
            # Look up the amino acid in the reverse mass table
            aa = reverse_mass_table.get(mass) # If the mass is not in the reverse mass table, return None
            if aa is not None:
                matching_peptides.append(aa) # Add the amino acid to the list of matching peptides
    return matching_peptides

if __name__ == "__main__":
    peptide = input("Enter the peptide: ")
    experimental_spectrum = list(map(int, input("Enter the experimental spectrum: ").split()))
    matching_peptides = compare_spectrum(theoretical_spectrum(peptide), experimental_spectrum)
    if matching_peptides:
        print("The matching peptides are: ", matching_peptides)
    else:
        print("No peptide with a theoretical spectrum matching the experimental spectrum exists.")
    
```
