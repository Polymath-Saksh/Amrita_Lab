```python
def transcription(main_strand):
    # return the complement of a DNA strand
    # join the complement of each base in the main strand

    complement = {'A': 'U', 'T': 'A', 'C': 'G', 'G': 'C'}
    #Step 1: Iterate through the main strand and get the complement of each base
    #Step 2: Join the complement of each base to form the complement strand

    return ''.join(complement[x] for x in main_strand)


def translation(rna_strand):
    # return the translation of an RNA strand
    # use a dictionary to map each codon to its corresponding amino acid
    # iterate through the RNA strand and get the amino acid for each codon
    # join the amino acids to form the protein

    #Step 1: Create a dictionary to map each codon to its corresponding amino acid
    #Step 2: Iterate through the RNA strand and get the amino acid for each codon
    #Step 3: Join the amino acids to form the protein

    codon_table = {
        "UUU": "F", "UUC": "F", "UUA": "L", "UUG": "L",
        "CUU": "L", "CUC": "L", "CUA": "L", "CUG": "L",
        "AUU": "I", "AUC": "I", "AUA": "I", "AUG": "M",
        "GUU": "V", "GUC": "V", "GUA": "V", "GUG": "V",
        "UCU": "S", "UCC": "S", "UCA": "S", "UCG": "S",
        "CCU": "P", "CCC": "P", "CCA": "P", "CCG": "P",
        "ACU": "T", "ACC": "T", "ACA": "T", "ACG": "T",
        "GCU": "A", "GCC": "A", "GCA": "A", "GCG": "A",
        "UAU": "Y", "UAC": "Y", "UAA": "*", "UAG": "*",
        "CAU": "H", "CAC": "H", "CAA": "Q", "CAG": "Q",
        "AAU": "N", "AAC": "N", "AAA": "K", "AAG": "K",
        "GAU": "D", "GAC": "D", "GAA": "E", "GAG": "E",
        "UGU": "C", "UGC": "C", "UGA": "*", "UGG": "W",
        "CGU": "R", "CGC": "R", "CGA": "R", "CGG": "R",
        "AGU": "S", "AGC": "S", "AGA": "R", "AGG": "R",
        "GGU": "G", "GGC": "G", "GGA": "G", "GGG": "G"
    }
    protein = ""

    if len(rna_strand) % 3 != 0:
        # If the length of the input strand is not a multiple of three, print an error message
        print("The length of the input strand must be a multiple of three for proper translation.")
    else:
        #Extract the codon from the RNA strand and get the corresponding amino acid
        for i in range(0, len(rna_strand), 3):
            codon = rna_strand[i:i+3]
            protein += codon_table[codon]
    return protein
    

def main():
    # get the main strand from the user
    main_strand = input("Enter the main strand: ")
    # get the RNA strand
    rna_strand = transcription(main_strand)
    print(f'Transcribed RNA: {rna_strand}')
    # get the protein
    protein = translation(rna_strand)
    # print the protein
    if protein != '':
        print(f'Protein: {protein}')

if __name__ == "__main__":
    main()

```
