## Filling <b>NA</b> with median

```python
df.info()
```

```python
df.isna().sum()
```

```python
new_df = df.copy()
```

```python
new_df=new_df.fillna(new_df[['Price','Distance','Landsize','Bathroom','Bedroom2','BuildingArea','YearBuilt']].median())
```

```python
new_df=new_df.fillna(new_df[['Car','Latitude','Longitude']].median())
```

```python
new_df.isna().sum()
```

## Drop the rows with other **NA** fields

```python
df = new_df
```

```python
df.dropna(inplace=True)
```
