```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('univ_rank.csv', encoding='latin-1')
```

```python
answers = dict()
```

```python
#Store df describe in answers
answers['1'] = df.describe()
```

```python
#Summarize the data
answers['2']=df.info()
```

```python
answers['3'] = df.tail(10)
```

```python
# Count of rows and columns
answers['4'] = df.shape
```

```python
answers['5'] = df['Research Quality'].unique()
```

```python
#Histogram for overall score and save it to answers['6]
plt.hist(df['Overall Score'])
plt.title('Overall Score')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.savefig('overall_score.png')

answers['6'] = plt.hist(df['Overall Score'])
```

```python
#Boxplot for name, overall score
plt.figure(figsize=(10,10))
sns.boxplot(x='Overall Score', y='Name', data=df)
plt.title('Overall Score')
plt.xlabel('Score')
plt.ylabel('University')
```
