### Boxplot (Seaborn)

```python
import seaborn as sns
sns.boxplot(data = df, x= 'area (km²)', y = '2023 population')
plt.title('2023 population vs area (km²)')
plt.xlabel('Area (km²)')
plt.ylabel('2023 population')
plt.show()
```

### Violinplot (Seaborn)

```python
# Violin plot with size adjusted to population
sns.violinplot(data = df, x= 'density (km²)', y = '1970 population' , density_norm='count')
plt.title('1970 population vs area (km²)')
plt.xlabel('Area (km²)')
plt.ylabel('1970 population')
plt.show()
```

### Scatterplot (Rank vs Continent)

```python
plt.scatter(x='continent' , y = 'rank', data = df)
plt.title('Rank vs Continent')
plt.xlabel('Continent')
plt.ylabel('Rank')
plt.show()
```

## Inconsistent Data Analysis and Handling

```python
df['continent'] = df['continent'].str.upper()
df['country'] = df['country'].str.upper()
```

```python
#Count occurences of each continent
print(df.continent.value_counts())
```

## Missing Data Analysis and Handling

```python
# Count missing values
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

### Outlier Handling based on z-score

```python
from scipy.stats import zscore
```

```python
z_scores = zscore(df['density (km²)'])
outliers = (z_scores > 3) | (z_scores < -3)
data_no_outliers = df[~outliers]    
```

```python
data_no_outliers['density (km²)'].hist()
```

## Handling Zero or Negative Values

```python
df['2020 population'] = df['2020 population'].apply(lambda x: max(x,0))
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.isnull().sum())
```

### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['area (km²)'] = scaler.fit_transform(df[['area (km²)']])
```

### Feature Engineering


```python
df['20cent population'] = df['2000 population'] + df['2010 population'] + df['2015 population'] + df['2020 population'] + df['2022 population'] + df['2023 population']
```

```python
df['20cent population'].hist()
plt.title('20th century population')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
