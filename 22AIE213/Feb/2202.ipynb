```python
df.rename(columns={'SepalLengthCm':'SepalLength','SepalWidthCm':'SepalWidth','PetalLengthCm':'PetalLength','PetalWidthCm':'PetalWidth'},inplace=True)
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(df,x='SepalLength',y='SepalWidth',hue='Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

## Pair Plots

```python
import warnings
warnings.filterwarnings("ignore")
sns.pairplot(df,hue='Species')
plt.show()
```

## Histogram of Features


```python
df.hist()
plt.show()
```

```python
df.drop('Id',axis=1,inplace=True)
```

## Kernel Density Estimation

```python
sns.kdeplot(df)
plt.title('KDE Plot')
plt.show()
```

```python
sns.violinplot(df,x="Species",y="PetalLength")
plt.title('Violin Plot of Petal Length By Species')
plt.show()
```

```python
sns.jointplot(df,x="PetalLength",y='PetalWidth',kind='reg')

plt.show()
```

```python
pd.plotting.radviz(df,'Species')
plt.title('Radviz Plot')
plt.show()
```

```python
sns.boxplot(df,x='Species',y='SepalLength')
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

```python
sns.kdeplot(df,x='SepalLength',y='SepalWidth',fill=True,color='cyan')
plt.title('Contour Plot of Sepal Length vs Sepal Width')
plt.show()
```

## Correlation HeatMap

```python
num_df = df.drop('Species',axis=1)
sns.heatmap(num_df.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
