```python
from hmmlearn import hmm
import numpy as np

obs_seq = np.array([[0],[1],[0],[0],[1]])

model = hmm.CategoricalHMM(n_components=2)

model.startprob_ = np.array([0.5, 0.5])
model.transmat_ = np.array([[0.9, 0.1],
                             [0.1, 0.9]])
model.emissionprob_ = np.array([[0.5,0.5],
                                [0.8, 0.2]])

model.fit(obs_seq)
hidden_states = model.predict(obs_seq)

print(f'Predicted Hidden States (0: Fair Coin, 1: Biased Coin):{hidden_states}')
```
