## Knapsack Problem

### Algorithm in words

- Given a list of items, each with a weight and a value, determine the number of each item to include in a collection so that the total weight is less than or equal to a given limit and the total value is as large as possible.

- The problem can be solved using dynamic programming. The idea is to create a 2D array `dp` of size `(n+1) x (W+1)`, where `n` is the number of items and `W` is the maximum weight that the knapsack can hold. The value of `dp[i][j]` is the maximum value that can be obtained with the first `i` items and a weight limit of `j`.

- The base case is when `i=0` or `j=0`, in which case the value is 0. Otherwise, if the weight of the `i`-th item is less than or equal to `j`, then the maximum value is the maximum of the value of the `i`-th item plus the value of the remaining items with a weight limit of `j` minus the weight of the `i`-th item, and the value of the remaining items with a weight limit of `j`. Otherwise, the maximum value is the value of the remaining items with a weight limit of `j`.

```python
import numpy as np

def knapsack(weights, values, capacity): #0/1 knapsack problem
    n = len(weights) # number of items
    dp = [[0]*(capacity+1) for _ in range(n+1)] # dp[i][j] is the maximum value of the first i items with capacity j

    for i in range(1, n+1): # i is the current item
        for j in range(1, capacity+1): # j is the current capacity
            if weights[i-1] <= j: # if the current item can be included
                dp[i][j] = max(values[i-1] + dp[i-1][j-weights[i-1]], dp[i-1][j])
            else: # if the current item cannot be included
                dp[i][j] = dp[i-1][j]

    arr_dp = np.array(dp)
    print(arr_dp)
    return dp[n][capacity]

def main():
    weights = [10, 20, 30]
    values = [60, 100, 120]
    capacity = 50

    print(knapsack(weights, values, capacity))

if __name__ == '__main__':
    main()
```

```python
# Recursive Solution

def knapsack_rec(weights, values, capacity, n=None):
    if n is None:
        n = len(weights) # number of items

    if (n == 0 or capacity == 0):
        return 0
    if (weights[n-1] > capacity): # if the current item cannot be included
        return knapsack_rec(weights, values, capacity,n-1)
    else: # if the current item can be included
        return max(
            values[n-1] + knapsack_rec(weights, values, capacity-weights[n-1], n-1),
            knapsack_rec(weights, values, capacity, n-1)
        )

def main1():
    weights = [10, 20, 30]
    values = [60, 100, 120]
    capacity = 50

    print(knapsack_rec(weights, values, capacity))

if __name__ == '__main__':
    main1()
```

### Stack Class Implementation

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, value):
        self.stack.append(value)
    
    def pop(self):
        if self.is_empty():
            return None
        self.stack = self.stack[:-1]
        return self.stack[-1]
    
    def peek(self):
        if self.is_empty():
            return None
        return self.stack[-1]
    
    def size(self):
        return len(self.stack)

    def is_empty(self):
        return len(self.stack) == 0
```

### Queue Class Implementation

```python
class Queue:
    def __init__(self):
        self.queue = []
    
    def is_empty(self):
        return len(self.queue) == 0

    def enqueue(self, value):
        self.queue.append(value)

    def dequeue(self):
        if self.is_empty():
            return None
        deq = self.queue[0]
        self.queue = self.queue[1:]
        return deq
    
    def size(self):
        return len(self.queue)
```

### Circular Queue Class Implementation

```python
class CircularQueue:
    def __init__(self, k):
        self.queue = [None]*k
        self.maxlen = k
        self.head = self.tail = None

    def enqueue(self, value):
        if self.is_full():
            return False
        if self.is_empty():
            self.head = self.tail = value
        else:
            self.tail = (self.tail + 1) % self.maxlen
            self.queue[self.tail] = value
        return True

    def dequeue(self):
        if self.is_empty():
            return False
        if self.head == self.tail:
            self.head = self.tail = None
        else:
            self.head = (self.head + 1) % self.maxlen
        return True

    def is_full(self):
        return self.head == (self.tail + 1) % self.maxlen

    def is_empty(self):
        return self.head is None
```
