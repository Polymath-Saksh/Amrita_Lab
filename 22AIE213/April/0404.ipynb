```python
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
iris_dat = load_iris()

X = iris_dat.data
y = iris_dat.target

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state = 446)
```

```python
std_scaler = StandardScaler()
X_train_scaled = std_scaler.fit_transform(X_train)
X_test_scaled = std_scaler.transform(X_test)
```

```python
svm_hard = SVC(kernel='linear', C=10000)
svm_hard.fit(X_train_scaled, y_train)
```

```python
svm_soft = SVC(kernel='linear', C=1)
svm_soft.fit(X_train_scaled, y_train)
```

```python
num_sv_hard = svm_hard.n_support_.sum()
print(num_sv_hard)
```

```python
num_sv_soft = svm_soft.n_support_.sum()
print(num_sv_soft)
```

```python
hard_margin_coef = svm_hard.coef_


print(f'Hard Margin SVC Coefficients\n')
for feature, coef in zip(iris_dat.feature_names, hard_margin_coef[0]):
    print(f"{feature}: {coef:.6f}")
```

```python
soft_margin_coef = svm_soft.coef_

print(f"Soft Margin SVM Coefficients:\n")
for feature, coef in zip(iris_dat.feature_names, soft_margin_coef[0]):
    print(f"{feature}: {coef:.6f}")
```

```python
print(f'Kernel Type(Hard SVM): {svm_hard.kernel}')
print(f'Kernel Type(Soft SVM): {svm_soft.kernel}')
```

```python
print(f'Kernel Type(Hard SVM): {svm_hard.kernel}')
print(f'Kernel Type(Soft SVM): {svm_soft.kernel}')
```

```python
svm_soft_rbf = SVC(kernel = 'rbf', C = 1)
svm_hard_rbf = SVC(kernel = 'rbf', C =10000)

svm_soft_rbf.fit(X_train_scaled, y_train)
svm_hard_rbf.fit(X_train_scaled, y_train)

num_sv_hard_rbf = svm_hard_rbf.n_support_.sum()
print("Hard Margins RBF: ",num_sv_hard_rbf)
num_sv_soft_rbf = svm_soft_rbf.n_support_.sum()
print("Soft Margins RBF: ",num_sv_soft_rbf)
```

```python
svm_soft_poly = SVC(kernel = 'poly', C = 1)
svm_hard_poly = SVC(kernel = 'poly', C =10000)

svm_soft_poly.fit(X_train_scaled, y_train)
svm_hard_poly.fit(X_train_scaled, y_train)

num_sv_hard_poly = svm_hard_poly.n_support_.sum()
print("Hard Margins Poly: ",num_sv_hard_poly)
num_sv_soft_poly = svm_soft_poly.n_support_.sum()
print("Soft Margins Poly: ",num_sv_soft_poly)
```

```python
svm_soft_sigmoid = SVC(kernel = 'sigmoid', C = 1)
svm_hard_sigmoid = SVC(kernel = 'sigmoid', C =10000)

svm_soft_sigmoid.fit(X_train_scaled, y_train)
svm_hard_sigmoid.fit(X_train_scaled, y_train)

num_sv_hard_sigmoid = svm_hard_sigmoid.n_support_.sum()
print("Hard Margins Sigmoid: ",num_sv_hard_sigmoid)
num_sv_soft_sigmoid = svm_soft_sigmoid.n_support_.sum()
print("Soft Margins Sigmoid: ",num_sv_soft_sigmoid)
```

```python
y_pred = svm_soft_sigmoid.predict(X_test_scaled)
misclassified_indices = [i for i, (true_label, pred_label) in enumerate(zip(y_test, y_pred)) if true_label != pred_label]
print("Misclassified instances in the SVM:")
for idx in misclassified_indices:
    print(f"Instance {idx + 1}: Predicted class {y_pred[idx]}, Actual class {y_test[idx]}")
```

```python
# Predictions using the soft margin SVM
y_pred_soft = svm_soft_sigmoid.predict(X_test_scaled)

# Calculate accuracy
accuracy_soft = accuracy_score(y_test, y_pred_soft)

print(f"Accuracy of the soft margin SVM: {accuracy_soft:.4f}")
```
