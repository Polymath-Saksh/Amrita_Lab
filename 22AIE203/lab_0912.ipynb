```python
from functools import reduce
def a_bar(L):
    # Use reduce to sum elements and divide by the count

    total, count = reduce(lambda acc, x: (acc[0] + x, acc[1] + 1), L, (0, 0))
    return total / count

def var(L, a_bar):
    # Use map and reduce to calculate variance
    diff_squared_sum, count = reduce(lambda acc, x: (acc[0] + (x - a_bar) ** 2, acc[1] + 1), L, (0, 0))
    return diff_squared_sum / count

if __name__ == "__main__":
    L = [1,2,3,-4,5]
    mean = a_bar(L)
    print(f'Mean is {mean}')
    variance = var(L, mean)
    print(f'Variance is {variance}')
```
