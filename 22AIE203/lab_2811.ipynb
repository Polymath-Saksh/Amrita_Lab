```python
from pydub import AudioSegment
from pydub.playback import play
from datetime import date
# for playing mp3 file


class Sound:
  def __init__(self):
    self.name = input("Enter the name: ")

  def birthday(self):
    for _ in range(5):
      print(f'Happy birthday to {self.name}')
    song = AudioSegment.from_mp3("/content/Nihal.mp3")
    play(song)

  def christmas(self):
    for _ in range(5):
      print(f'Merry Christmas {self.name}!!')
    song = AudioSegment.from_mp3("/content/Christmas.mp3")
    play(song)

  def diwali(self):
    for _ in range(5):
      print(f'Happy Diwali {self.name}!!')
    song = AudioSegment.from_mp3("/content/Diwali.mp3")
    play(song)

  def newyear(self):
    for _ in range(5):
      print(f'Happy New Year {self.name}!!')
    song = AudioSegment.from_mp3("/content/NewYear.mp3")
    play(song)

if __name__=='__main__':
  process = Sound()
  today_date = date.today()
  if today_date.month == '12' and today_date.day=='25':
    process.christmas()
  elif today_date.month =='1' and today_date.day=='1':
    process.newyear()
  elif today_date.month =='11' and today_date.day=='12':
    process.diwali()
  else:
    process.birthday()
```

```python
class Shopping:
    def __init__(self):
        self.cart = {}

    def add_item(self, item, cost, quantity):
        if item in self.cart:
            # If the item is already in the cart, update the quantity
            self.cart[item][1] += quantity
        else:
            # If the item is not in the cart, add it with the cost and quantity
            self.cart[item] = [cost, quantity]
        print(f"{quantity} {item}(s) added to the cart.")

    def remove_item(self, item, quantity):
        if item in self.cart:
            if quantity >= self.cart[item][1]:
                # If the quantity to remove is greater than or equal to the quantity in the cart, remove the item
                del self.cart[item]
            else:
                # Otherwise, update the quantity in the cart
                self.cart[item][1] -= quantity
            print(f"{quantity} {item}(s) removed from the cart.")
        else:
            print(f"{item} is not in the cart.")

    def view_cart(self):
        if not self.cart:
            print("Your cart is empty.")
        else:
            print("Cart Contents:")
            for item, details in self.cart.items():
                cost, quantity = details
                print(f"{item}: {quantity} at ${cost} each")

    def print_bill(self):
        total_cost = sum(cost * quantity for cost, quantity in self.cart.values())
        if not self.cart:
            print("Your cart is empty. No bill generated.")
        else:
            print("------ Bill ------")
            for item, details in self.cart.items():
                cost, quantity = details
                print(f"{item}: {quantity} at ${cost} each")
            print(f"Total: ${total_cost}")
            print("Thank you for shopping with us!")

# Example usage:
def main():
    shopping_cart = Shopping()

    while True:
        print("\n--- Shopping Menu ---")
        print("1. Add Item to Cart")
        print("2. Remove Item from Cart")
        print("3. View Cart")
        print("4. Print Bill")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == "1":
            item = input("Enter item name: ")
            cost = float(input("Enter item cost: "))
            quantity = int(input("Enter quantity: "))
            shopping_cart.add_item(item, cost, quantity)

        elif choice == "2":
            item = input("Enter item name to remove: ")
            quantity = int(input("Enter quantity to remove: "))
            shopping_cart.remove_item(item, quantity)

        elif choice == "3":
            shopping_cart.view_cart()

        elif choice == "4":
            shopping_cart.print_bill()

        elif choice == "5":
            print("Exiting. Thank you for shopping!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

if __name__ == "__main__":
    main()
```
