```python
def counting_sort(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 10

    for i in range(n):
        index = arr[i] // exp
        count[index % 10] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]

    i = n - 1
    while i >= 0:
        index = arr[i] // exp
        output[count[index % 10] - 1] = arr[i]
        count[index % 10] -= 1
        i -= 1

    for i in range(n):
        arr[i] = output[i]

def radix_sort(arr):
    max_value = max(arr)
    exp = 1

    while max_value // exp > 0:
        counting_sort(arr, exp)
        exp *= 10

    return arr

# Example usage:
arr = [170, 45, 75, 90, 802, 24, 2, 66]
sorted_arr = radix_sort(arr)
print(sorted_arr)
```

```python
from colorama import Fore, Back, Style

def change_foreground_color(color):
    return lambda text: color + text + Style.RESET_ALL

def change_background_color(color):
    return lambda text: color + text + Style.RESET_ALL

def change_style(style):
    return lambda text: style + text + Style.RESET_ALL

def apply_formatting(foreground_func, background_func, style_func, text):
    formatted_text = foreground_func(background_func(style_func(text)))
    return formatted_text

# Example usage:
formatted_text = apply_formatting(change_foreground_color(RED), change_background_color(Back.WHITE), change_style(Style.BRIGHT), "Hello, world!")
print(formatted_text)
```
