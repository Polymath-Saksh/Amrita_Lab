## Binary Search (Iterative)

```python
def binary(arr,n,ele):
    low,high=0,n-1
    
    while high>low:

        mid = (low+high)//2
        if arr[mid] == ele:
            return mid+1
        elif arr[mid]>ele:
            high=mid
        else:
            low=mid

    return -1


if __name__=='__main__':
    arr=[]
    n= int(input("Enter size of array: "))
    for _ in range(n):
        dat=int(input())
        arr.append(dat)
    arr=sorted(arr)

    ele = int(input("Enter element to be search: "))
    bin_val=binary(arr,n,ele)

    if bin_val !=-1:
        print(f'The element is at index {bin_val}')
    else:
        print(f'The element is not present.')
```

## Binary Search (Recursive)

```python
# Binary Recursive

def binary_rec(arr,ele,low,high):
    if low>high:
        return -1
    mid = (low+high)//2

    if arr[mid]==ele:
        return mid+1
    elif arr[mid]>ele:
        return binary_rec(arr,ele,low,mid-1)
    else:
        return binary_rec(arr,ele,mid+1,high)


if __name__=='__main__':
    arr=[]
    n= int(input("Enter size of array: "))
    for _ in range(n):
        dat=int(input())
        arr.append(dat)
    arr=sorted(arr)

    ele = int(input("Enter element to be search: "))
    bin_val=binary_rec(arr,ele,0,n-1)

    if bin_val !=-1:
        print(f'The element is at index {bin_val}')
    else:
        print(f'The element is not present.')
```

## Insertion Sort

```python
def insertion_sort(arr,n):
    for i in range(1,n):
        key = arr[i]
        j=i-1
        while j>=0 and key<arr[j]:
            arr[j+1] = arr[j]
            j-=1
            arr[j+1] = key
    return arr

if __name__=='__main__':
    n = int(input("Enter the number of elements: "))
    arr=[]
    for i in range(n):
        arr.append(int(input()))
    print(f'The sorted elements are {insertion_sort(arr,n)}')
```

## Bubble Sort

```python
def bubble_sort(arr,n):
    for i in range(n):
        for j in range(i+1,n):
            if arr[j]<arr[j-1]:
                arr[j],arr[j-1]=arr[j-1],arr[j]
    
    return arr

if __name__=='__main__':
    n = int(input("Enter the number of elements: "))
    arr=[]
    for i in range(n):
        arr.append(int(input()))
    print(f'The sorted elements are {bubble_sort(arr,n)}')
```
