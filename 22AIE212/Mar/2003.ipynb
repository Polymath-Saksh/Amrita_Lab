## Huffman Encoding

```python
import heapq

class Node:
    def __init__(self, symbol, prob, left = None, right = None):
        self.symbol = symbol
        self.prob = prob
        self.left = left
        self.right = right
        self.huff = ''

    def __lt__(self, next):
        return self.prob < next.prob
    
def printNodes(node, val =''):
    newVal = val + str(node.huff)

    if(node.left):
        printNodes(node.left,newVal)
    if(node.right):
        printNodes(node.right, newVal)

    if (not node.left and not node.right):
        print(f'{node.symbol} -> {newVal}')

def huffman_coding(chars, freq):
    nodes = []

    for x in range(len(chars)):
        heapq.heappush(nodes, Node(chars[x], freq[x]))

    while len(nodes) > 1:
        left = heapq.heappop(nodes)
        right = heapq.heappop(nodes)

        left.huff = 0
        right.huff = 1

        newNode = Node(None, left.prob + right.prob, left, right)
        heapq.heappush(nodes, newNode)

    printNodes(nodes[0])
```

### AMRITAAI

```python

chars = ['A','M','R','I','T']
freq = [3,1,1,1,1]
huffman_coding(chars, freq)
```

### AMARAVATIAI

```python

chars = ['A','M','R','V','T','I']
freq = [5,2,1,1,1,1]
huffman_coding(chars, freq)
```
