### Max Heap

```python
def heapify(arr,n,i): #i 
    largest = i
    l, r = 2*i + 1, 2*i + 2
    if l < n and arr[largest] < arr[l]:
        largest = l
    if r < n and arr[largest] < arr[r]:
        largest = r
    if largest != i: # if largest is not the root
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest) # recursively heapify the affected sub-tree

def heapsort(arr):
    n = len(arr)
    for i in range(n//2 - 1, -1, -1): # build max heap
        heapify(arr, n, i)
    for i in range(n-1, -1 , -1): # extract elements
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)
    return arr

print(heapsort([3,8,1,6,33,90]))
```

### Quick Sort

```python
def partition(arr,low, high):
    pivot = arr[high]

    i = low - 1

    for j in range(low, high): # iterate through the array
        if arr[j] <= pivot: # if current element is smaller than pivot
            i+=1
            arr[i], arr[j] = arr[j], arr[i] 
    arr[i+1], arr[high] = arr[high], arr[i+1]
        
    return i+1

def quickSort(arr, low, high):
    if low < high:
        pivot = partition(arr,low,high)
        quickSort(arr,low, pivot-1)
        quickSort(arr,pivot+1, high)


if __name__ == '__main__':
    array = [10, 7, 8, 9, 1, 5]
    N = len(array)

    # Function call
    quickSort(array, 0, N - 1)
    print('Sorted array:')
    for x in array:
        print(x, end=" ")
```

### Merge Sort

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]
        merge_sort(L)
        merge_sort(R)
        i = j = k = 0

        while i<len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i+=1
            else:
                arr[k] = R[j]
                j+=1
            k+=1

        while i < len(L):
            arr[k] = L[i]
            i+=1
            k+=1
        while j < len(R):
            arr[k] = R[j]
            j+=1
            k+=1

        return arr

if __name__ == '__main__':
    arr = [12, 11, 13, 5, 6, 7]
    print('Given array is', arr)
    print('Sorted array is', merge_sort(arr))
```

```python
#Insertion sort

def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1 #previous element
        while j >= 0 and key < arr[j]: #key is less than previous element
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

if __name__ == '__main__':
    arr = [12, 11, 13, 5, 6]
    print('Given array is', arr)
    print('Sorted array is', insertion_sort(arr))
```

### Fractional Knapsack

```python
def fractional_knapsack(W,V,C):
    ratios = []
    for i in range(len(W)):
        ratios.append((V[i]/W[i],W[i],V[i]))
    ratios.sort(reverse=True)

    max_value = 0
    for ratio in ratios:
        if C == 0:
            break
        if ratio[1] <= C: #if weight is less than capacity
            max_value += ratio[2] #add whole value to max_value
            C -= ratio[1]
        else: #if weight is greater than capacity
            max_value += ratio[0] * C #add fraction of value to max_value
            C = 0
    return max_value

if __name__ == '__main__':
    val = [60, 100, 120]
    wt = [10, 20, 30]
    W = 50
    print('Maximum value:', fractional_knapsack(wt,val,W))
```

### 0/1 Knapsack (DP Approach)

```python
def knapsack_0_1(W, wt, val, n):
    dp = [[0 for _ in range(W+1)] for _ in range(n+1)] #initialize dp array with 0 of size n+1 * W+1
    for i in range(1, n+1):
        for j in range(1, W+1):
            if wt[i-1] <= j: #if weight of current item is less than or equal to current capacity
                dp[i][j] = max(val[i-1] + dp[i-1][j - wt[i-1]], dp[i-1][j])
            else:
                dp[i][j] = dp[i-1][j]

    return dp[n][W] #Last element of dp array will be the maximum value

def main():
    val = [60, 100, 120]
    wt = [10, 20, 30]
    W = 50
    n = len(val)
    print(knapsack_0_1(W, wt, val, n))

val = [60, 100, 120]
wt = [10, 20, 30]
W = 50
n = len(val)
print(knapsack_0_1(W, wt, val, n))
```

### Huffman Encoding

```python
import heapq

class Node:
    def __init__(self, freq, sym, left= None, right=None):
        self.freq = freq
        self.sym = sym
        self.left = left
        self.right = right
        self.huff = '' #Tree Direction

    def __lt__(self, next): # Work : Compare the frequency of the nodes
        return self.freq < next.freq

def printNodes(node, val = ''):
    
    newVal = val + str(node.huff)

    if(node.left):
        printNodes(node.left, newVal)
    if(node.right):
        printNodes(node.right,newVal)

    if (not node.right and not node.left): #If Edge Node
        print(f'{node.sym} -> {newVal}')
chars = ['S','A','K','H','M']
freq = [2,2,1,1,1]

def huffman(chars, freq):
    nodes = [] #list of unused nodes

    for x in range(len(chars)):
        heapq.heappush(nodes, Node(freq[x], chars[x]))

    while len(nodes) > 1:
        left = heapq.heappop(nodes)
        right = heapq.heappop(nodes)

        left.huff, right.huff = 0,1

        newNode = Node(left.freq + right.freq, left.sym + right.sym, left, right)

        heapq.heappush(nodes,newNode)


    printNodes(nodes[0])

huffman(chars, freq)
```

## Kruskal Algorithm

```python
class Graph:
    def __init__(self, vertices):
        self.v = vertices
        self.graph = []

    def addEdge(self, u, v, w):
        self.graph.append([u, v, w])

    def find(self, parent, i):  # Common parent function
        if parent[i] != i: # if parent is not same as i
            parent[i] = self.find(parent, parent[i])  # path compression by setting parent of i to root
        return parent[i]  # return root if parent is same as i

    def union(self, parent, rank, x, y):  # Union function merges two sets
        xroot = self.find(parent, x)
        yroot = self.find(parent, y)
        if rank[xroot] < rank[yroot]:
            parent[xroot] = yroot
        elif rank[xroot] > rank[yroot]:
            parent[yroot] = xroot
        else:
            parent[yroot] = xroot
            rank[xroot] += 1  # increment rank if both ranks are same

    def kruskal(self):
        result = []  # This will store the resultant MST
        i = e = 0  # index variables for sorted edges
        self.graph = sorted(self.graph, key=lambda x: x[2])  # sort edges by weight

        parent = []
        rank = []

        for node in range(self.v):  # Initialize parent and rank arrays
            parent.append(node)
            rank.append(0)

        while e < self.v - 1:
            u, v, w = self.graph[i]
            i += 1
            x = self.find(parent, u)
            y = self.find(parent, v)

            if x != y:  # If including this edge doesn't cause cycle
                e += 1
                result.append([u, v, w])
                self.union(parent, rank, x, y)

        print('Edges in the MST:')
        for u, v, w in result:
            print(f'{u} - {v} -> {w}')

g = Graph(4)
g.addEdge(0, 1, 10)
g.addEdge(0, 2, 6)
g.addEdge(0, 3, 5)
g.addEdge(1, 3, 15)
g.addEdge(2, 3, 4)
g.kruskal()
```

### Prim's Algorithm

```python
class Graph:
    def __init__(self, vertices):
        self.v = vertices
        self.graph = [[0 for _ in range(vertices)] for _ in range(vertices)]

    def printMST(self, parent):
        print('Edge \tWeight')
        for i in range(1, self.v):
            print(f'{parent[i]} - {i} \t{self.graph[i][parent[i]]}')

    def minKey(self, key, mstSet): #mstSet is a set of vertices not yet included in MST, key is the key values of vertices
        min = float('inf')
        for v in range(self.v): # Find the vertex with minimum key value from the set of vertices not yet included in MST
            if key[v] < min and mstSet[v] == False:  # Pick the vertex with minimum key value
                min = key[v]
                min_index = v
        return min_index

    def primMST(self):
        key = [float('inf')] * self.v # Key values used to pick minimum weight edge in cut
        parent = [None] * self.v #Array to store constructed MST
        key[0] = 0
        mstSet = [False] * self.v
        parent[0] = -1

        for _ in range(self.v): # Construct MST
            u = self.minKey(key, mstSet)
            mstSet[u] = True
            for v in range(self.v): # Update key value and parent index of the adjacent vertices of the picked vertex
                if self.graph[u][v] > 0 and mstSet[v] == False and key[v] > self.graph[u][v]: # Update the key only if graph[u][v] is smaller than key[v]
                    key[v] = self.graph[u][v]
                    parent[v] = u
        self.printMST(parent)

g = Graph(5)
g.graph = [[0, 2, 0, 6, 0],
           [2, 0, 3, 8, 5],
           [0, 3, 0, 0, 7],
           [6, 8, 0, 0, 9],
           [0, 5, 7, 9, 0]] #Rows are vertices and columns are edges hence it is a 5x5 weighted adjacency matrix
g.primMST()
```
