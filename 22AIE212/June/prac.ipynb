## Knapsack Problem

### Algorithm in words

- Given a list of items, each with a weight and a value, determine the number of each item to include in a collection so that the total weight is less than or equal to a given limit and the total value is as large as possible.

- The problem can be solved using dynamic programming. The idea is to create a 2D array `dp` of size `(n+1) x (W+1)`, where `n` is the number of items and `W` is the maximum weight that the knapsack can hold. The value of `dp[i][j]` is the maximum value that can be obtained with the first `i` items and a weight limit of `j`.

- The base case is when `i=0` or `j=0`, in which case the value is 0. Otherwise, if the weight of the `i`-th item is less than or equal to `j`, then the maximum value is the maximum of the value of the `i`-th item plus the value of the remaining items with a weight limit of `j` minus the weight of the `i`-th item, and the value of the remaining items with a weight limit of `j`. Otherwise, the maximum value is the value of the remaining items with a weight limit of `j`.

- Procedure (in verbal form)

```
- Create a 2D array called `dp` with dimensions `(n+1) x (W+1)` and initialize all elements to 0.
- Iterate over each item from 1 to `n`:
  - For each item, iterate over each weight from 1 to `W`:
    - If the weight of the current item is less than or equal to the current weight limit `j`:
      - Set `dp[i][j]` to the maximum value between the value of the current item plus the value of the remaining items with a weight limit of `j` minus the weight of the current item (`dp[i-1][j-weight[i-1]]`) and the value of the remaining items with a weight limit of `j` (`dp[i-1][j]`).
    - Otherwise, set `dp[i][j]` to the value of the remaining items with a weight limit of `j` (`dp[i-1][j]`).
- Return the value at `dp[n][W]`, which represents the maximum value that can be obtained with the given items and weight limit.

```

````python
import numpy as np

def knapsack(weights, values, capacity): #0/1 knapsack problem
    n = len(weights) # number of items
    dp = [[0]*(capacity+1) for _ in range(n+1)] # dp[i][j] is the maximum value of the first i items with capacity j

    for i in range(1, n+1): # i is the current item
        for j in range(1, capacity+1): # j is the current capacity
            if weights[i-1] <= j: # if the current item can be included
                dp[i][j] = max(values[i-1] + dp[i-1][j-weights[i-1]], dp[i-1][j])
            else: # if the current item cannot be included
                dp[i][j] = dp[i-1][j]

    dp_n = np.array(dp) 
    print(dp_n)
    return dp[n][capacity]

def main():
    weights = [10, 20, 30]
    values = [60, 100, 120]
    capacity = 50

    print(knapsack(weights, values, capacity))

if __name__ == '__main__':
    main()
````

````python
# Recursive Solution
def knapsack_rec(weights, values, capacity, n=None):
    if n is None:
        n = len(weights) # number of items

    if (n == 0 or capacity == 0):
        return 0
    if (weights[n-1] > capacity): # if the current item cannot be included
        return knapsack_rec(weights, values, capacity,n-1)
    else: # if the current item can be included
        return max(
            values[n-1] + knapsack_rec(weights, values, capacity-weights[n-1], n-1),
            knapsack_rec(weights, values, capacity, n-1)
        )

def main1():
    weights = [10, 20, 30]
    values = [60, 100, 120]
    capacity = 50

    print(knapsack_rec(weights, values, capacity))

if __name__ == '__main__':
    main1()
````

## Floyd Warshall Algorithm

````python
def floyd_warshall(graph):
    n = len(graph)

    dist = [ row[:] for row in graph]

    for k in range(n):
        for i in range(n):
            for j in range(n):
                int_dist = dist[i][k] + dist[k][j]
                if dist[i][j] > int_dist:
                    dist[i][j] = int_dist
    return dist

graph = [[0,5,float('inf'), 10],
            [float('inf'), 0, 3, float('inf')],
            [float('inf'), float('inf'), 0,   1],
            [float('inf'), float('inf'), float('inf'), 0]
            ]

print(floyd_warshall(graph))


````

### Stack Class Implementation

````python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, value):
        self.stack.append(value)
    
    def pop(self):
        if self.is_empty():
            return None
        self.stack = self.stack[:-1]
        return self.stack[-1]
    
    def peek(self):
        if self.is_empty():
            return None
        return self.stack[-1]
    
    def size(self):
        return len(self.stack)

    def is_empty(self):
        return len(self.stack) == 0
````

### Queue Class Implementation

````python
class Queue:
    def __init__(self):
        self.queue = []
    
    def is_empty(self):
        return len(self.queue) == 0

    def enqueue(self, value):
        self.queue.append(value)

    def dequeue(self):
        if self.is_empty():
            return None
        deq = self.queue[0]
        self.queue = self.queue[1:]
        return deq
    
    def size(self):
        return len(self.queue)
````

### Circular Queue Class Implementation

````python
class CircularQueue:
    def __init__(self, k):
        self.queue = [None]*k
        self.maxlen = k
        self.head = self.tail = None

    def enqueue(self, value):
        if self.is_full():
            return False
        if self.is_empty():
            self.head = self.tail = value
        else:
            self.tail = (self.tail + 1) % self.maxlen
            self.queue[self.tail] = value
        return True

    def dequeue(self):
        if self.is_empty():
            return False
        if self.head == self.tail:
            self.head = self.tail = None
        else:
            self.head = (self.head + 1) % self.maxlen
        return True

    def is_full(self):
        return self.head == (self.tail + 1) % self.maxlen

    def is_empty(self):
        return self.head is None
````

## Merge Sort

````python
def merge_sort(arr):
    if len(arr)>1:
        mid = len(arr)//2 #Finding the mid of the array
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)
        i = j = k = 0 # i is the index of L, j is the index of R, k is the index of arr

        while i<len(L) and j < len(R): 
            if L[i] > R[j]: # if the element in L is greater than the element in R
                arr[k] = R[j]
                j+=1
            else: # if the element in L is less than the element in R
                arr[k] = L[i]
                i+=1
            k+=1

        while i < len(L):# if there are any remaining elements in L
            arr[k] = L[i]
            i+=1
            k+=1

        while j < len(R):# if there are any remaining elements in R
            arr[k] = R[j]
            j+=1
            k+=1

def main3():
    arr = [12, 11, 13, 5, 6, 7]
    merge_sort(arr)
    print(arr)

main3()
````

### Recursive Binary Search

````python
def bin_search_rec(arr, ele, low, high):
    if low > high :
        return -1

    mid = (low + high) //2

    if arr[mid] == ele:
        return mid+1

    if arr[mid] > ele:
        return bin_search_rec(arr, ele, low, mid-1) # search in the left half

    if arr[mid] < ele:
        return bin_search_rec(arr, ele, mid+1, high) # search in the right half

def main4():
    arr = [2, 3, 4, 10, 40]
    ele = 10
    print(f'\'{ele}\' found at index {bin_search_rec(arr, ele, 0, len(arr)-1)}')

main4()
````

### Insertion Sort

- The algorithm works by dividing the input list into two parts: a sorted part and an unsorted part. Initially, the sorted part is empty and the unsorted part is the entire list. The algorithm then iterates over the unsorted part, removing one element at a time and inserting it into the correct position in the sorted part. This process is repeated until the unsorted part is empty.

- Procedure

    1. Start with the first element of the list as the sorted part.
    2. Iterate over the unsorted part of the list.
    3. Remove the first element of the unsorted part.
    4. Insert the removed element into the correct position in the sorted part.
    5. Repeat steps 2-4 until the unsorted part is empty.

````python
def insertion_sort(arr):
    for i in range(1,len(arr)): #pass through the array starting from the second element
        key = arr[i]
        j = i-1 # index of the element before the key
        while j>=0 and key < arr[j]: # if the key is less than the element before it, it means it has to be placed before it, so it is ignoring the elements that are greater than the key by backtracking
            arr[j+1] = arr[j] # move the element to the right which is greater than the key to make space for the key
            j-=1
        arr[j+1] = key # place the key in the correct position
    return arr

def main5():
    arr = [12, 11, 13, 5, 6]
    print(insertion_sort(arr))

main5()
````

````python
def sin_series(x, n):
    sum = 0
    for i in range(n): # n is the number of terms
        term = (-1)**i * x**(2*i+1) / math.factorial(2*i+1)
        sum += term
    return sum

def cos_series(x, n):
    sum = 0
    for i in range(n):
        term = (-1)**i * x**(2*i) / math.factorial(2*i)
        sum += term
    return sum

def tan_series(x, n):
    return sin_series(x, n) / cos_series(x, n)

def main5():
    x = 30
    n = 20
    print(f'Sine of {x} is {sin_series(math.radians(x), n)}')
    print(f'Cosine of {x} is {cos_series(math.radians(x), n)}')
    print(f'Tangent of {x} is {tan_series(math.radians(x), n)}')

main5()
````

## Radix Sort

````python
def counting_sort(arr,exp): #exp is the current digit
    n = len(arr)
    output = [0]*n
    count = [0]*10 # count array to store the count of each element

    for i in range(n): # count the number of occurrences of each element
        index = arr[i]//exp # index is the current digit
        count[index%10]+=1

    for i in range(1,10): # calculate the position of each element in the output array
        count[i]+=count[i-1]

    i = n-1
    while i>=0:
        index = arr[i]//exp # index is the current digit
        output[count[index%10]-1] = arr[i] # place the element in the correct position
        count[index%10]-=1 # decrement the count of the element
        i-=1

    for i in range(n):
        arr[i] = output[i]

def radix_sort(arr):
    max1 = max(arr)
    exp = 1
    while max1//exp > 0:
        counting_sort(arr,exp)
        exp*=10
    return arr

def main6():
    arr = [170, 45, 75, 90, 802, 24, 2, 66]
    radix_sort(arr)
    print(arr)

main6()
````

````python
def heapify(arr,n,i): # n is the size of the heap, i is the index of the root
    largest = i
    l,r = 2*i+1,2*i+2

    if l<n and arr[largest] < arr[l]: # if the left child is greater than the root
        largest = l

    if r<n and arr[largest] < arr[r]: # if the right child is greater than the root
        largest = r

    if largest != i: # if the largest is not the root
        arr[i],arr[largest] = arr[largest],arr[i] # swap
        heapify(arr,n,largest)

def heap_sort(arr):
    n = len(arr)
    for i in range(n//2-1,-1,-1): # build a max heap
        heapify(arr,n,i)

    for i in range(n-1,0,-1):
        arr[i],arr[0] = arr[0],arr[i] # swap
        heapify(arr,i,0)

def main7():
    arr = [12, 11, 13, 5, 6, 7]
    heap_sort(arr)
    print(arr)

main7()
````

## Huffmann Encoding

````python
import heapq

class Node:
    def __init__(self, sym, prob, left=None, right=None):
        self.sym = sym # symbol
        self.prob = prob # probability
        self.left = left # left child
        self.right = right # right child
        self.huff = '' # huffman code

    def __lt__(self, nxt):
        return self.prob < nxt.prob

def printNodes(node, val = ''):
    newVal = val + str(node.huff)
    if node.left:
        printNodes(node.left, newVal)
    if node.right:
        printNodes(node.right, newVal)

    if not node.left and not node.right:
        print(f'{node.sym} : {newVal}')

def huffman_tree(chars, freq):
    nodes = []

    for i in range(len(chars)): # create a node for each character
        heapq.heappush(nodes,(Node(chars[i], freq[i])))

    while len(nodes) > 1:
        left = heapq.heappop(nodes) # pop the two nodes with the lowest probability
        right = heapq.heappop(nodes)# pop the two nodes with the lowest probability

        left.huff = '1' # assign 0 to the left child
        right.huff = '0' # assign 1 to the right child

        newNode = Node(left.sym+right.sym, left.prob+right.prob, left, right) # create a new node with the sum of the probabilities of the two nodes
        heapq.heappush(nodes, newNode) # push the new node into the heap

    printNodes(nodes[0])

def main8():
    chars = ['A','M','R','I','T']
    freq = [3,1,1,2,1]

    huffman_tree(chars, freq)

main8()
````
