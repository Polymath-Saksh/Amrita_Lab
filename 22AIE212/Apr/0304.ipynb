## Knapsack Problem (0/1)

```python
def knapsack(items, capacity):
    n = len(items)
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)] # dp[i][j] is the maximum value that can be obtained with the first i items and a capacity of j
    for i in range(1, n + 1): # i is the number of items
        for j in range(1, capacity + 1): # `j` is the capacity
            if items[i - 1][1] > j: # if the weight of the current item is more than the current capacity, we can't include it
                dp[i][j] = dp[i - 1][j] # so the maximum value that can be obtained with the first `i` items and a capacity of `j` is the same as the maximum value that can be obtained with the first `i - 1` items and a capacity of `j`
            else:# otherwise, we must choose between including the current item or excluding it
                dp[i][j] = max(dp[i - 1][j], dp[i - 1][j - items[i - 1][1]] + items[i - 1][0]) # the maximum value that can be obtained with the first `i` items and a capacity of `j` is the maximum of the maximum value that can be obtained with the first `i - 1` items and a capacity of `j` and the maximum value that can be obtained with the first `i - 1` items and a capacity of `j - items[i - 1][1]` plus the value of the current item
    return dp[n][capacity]# the maximum value that can be obtained with all the items and the full capacity
```

## All pairs shortest path problem

```python
def allpairs(items):
    n = len(items)
    dp = [[0 for _ in range(n)] for _ in range(n)] # dp[i][j] is the maximum value that can be obtained with the first i items and a capacity of j
    for i in range(1, n): # i is the number of items
        for j in range(i): # `j` is the capacity
            dp[i][j] = knapsack(items[:i], items[j][1]) # the maximum value that can be obtained with the first `i` items and a capacity of `j` is the maximum value that can be obtained with the first `i` items and a capacity of `j`
    return dp

def main():
    items = [(60, 10), (100, 20), (120, 30)]
    print(knapsack(items, 50))
    
    print(allpairs(items))
```

```python
main()
```
