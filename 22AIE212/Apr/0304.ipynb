## Knapsack Problem (0/1)

```python
def knapsack(items, capacity):
    n = len(items)
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)] 
    for i in range(1, n + 1):
        for j in range(1, capacity + 1):
            if items[i - 1][1] > j:
                dp[i][j] = dp[i - 1][j] 
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i - 1][j - items[i - 1][1]] + items[i - 1][0])
    return dp[n][capacity]
```

## All pairs shortest path problem

```python
def allpairs(items):
    n = len(items)
    dp = [[0 for _ in range(n)] for _ in range(n)] # dp[i][j] is the maximum value that can be obtained with the first i items and a capacity of j
    for i in range(1, n): # i is the number of items
        for j in range(i): # `j` is the capacity
            dp[i][j] = knapsack(items[:i], items[j][1]) # the maximum value that can be obtained with the first `i` items and a capacity of `j` is the maximum value that can be obtained with the first `i` items and a capacity of `j`
    return dp

def main():
    items = [(60, 10), (100, 20), (120, 30)]
    print(knapsack(items, 100))
    print(allpairs(items))
```

```python
main()
```

```python
def all_pair(W,A):
    n = len(A)
    A[0] = [0 for _ in range(n)]
    for i in range(1,n):
        for j in range(n):
            A[i][j] = W[i][j]
    for k in range(1,n):
        for i in range(1,n):
            for j in range(1,n):
                A[i][j] = min(A[i][j],A[i][k]+A[k][j])
    return A

def main():
    W = [[0, 3, 8, float('inf'), -4],
         [float('inf'), 0, float('inf'), 1, 7],
         [float('inf'), 4, 0, float('inf'), float('inf')],
         [2, float('inf'), -5, 0, float('inf')],
         [float('inf'), float('inf'), float('inf'), 6, 0]]
    A = [[0 for _ in range(5)] for _ in range(5)]
    print(all_pair(W,A))
    

main()
```
