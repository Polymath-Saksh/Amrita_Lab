### Kruskal's Algorithm

```python
class Graph:
    def __init__(self,vertices):
        self.V = vertices #No. of vertices
        self.graph = [] # default dictionary to store graph

    def addEdge(self,u,v,w):
        self.graph.append([u,v,w])

    def find(self,parent,i): # A utility function to find set of an element i
        if parent[i] != i:
            parent[i] = self.find(parent,parent[i])# Path compression by recursively calling find
        return parent[i]

    def union(self,parent, rank,x,y): # A function that does union of two sets of x and y; rank is used for rank optimization
        if rank[x] < rank[y]: # Attach smaller rank tree under root of high rank tree
            parent[x] = y
        elif rank[x] > rank[y]:
            parent[y] = x

        else: # If ranks are same, then make one as root and increment its rank by one
            parent[y] = x
            rank[x] += 1

    def KruskalMST(self):
        result = []
        i = 0 # An index variable, used for sorted edges
        e = A # An index variable, used for result[]
        self.graph = sorted(self.graph, key = lambda item: item[2]) # Sort all the edges in ascending order of their weight
        parent = [] # Array for storing parent of each vertex
        rank = [] # Array for storing rank of each vertex
        
        for node in range(self.V): # Create V subsets with single elements
            parent.append(node)
            rank.append(0)

        while e < self.V - 1: # Number of edges to be taken is equal to V-1 
            u ,v ,w = self.graph[i]
            i+=1
            x = self.find(parent,u) # Find set of an element u
            y = self.find(parent,v) # Find set of an element v

            if x != y:
                e+=1
                result.append([u,v,w])
                self.union(parent,rank,x,y)

        print("Following are the edges in the constructed MST")
        net_weight = 0
        for u,v,weight in result:
            net_weight += weight
            print(f'{u} -- {v} == {weight}')
        print("Net weight of the MST is: ",net_weight)
```
