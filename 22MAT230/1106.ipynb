### Fourier Transform Matrix

```python
import numpy as np

def fourier_matrix(n):
    F = np.zeros((n,n), dtype = complex)
    omega = np.exp((-2j*np.pi)/n)
    for i in range(n):
        for j in range(n):
            if i == 0 or j == 0:
                F[i,j] = 1
            else:
                F[i,j] = omega**(i*j)
    return F

if __name__ == "__main__":
    main()
```

### Discrete Fourier Transform

```python
C = np.array([1, 0, 1, 0], dtype = complex)
F = fourier_matrix(4)
DFT = np.dot(F, C)

print(DFT)
```

### Fast Fourier Transform

```python
# Fast Fourier Transform 

def fft(x):
    n = len(x)
    if n == 1:
        return x
    else:
        even = fft(x[::2])
        odd = fft(x[1::2])
        factor = np.exp(-2j*np.pi*np.arange(n)/n)
        return np.concatenate([even + factor[:n//2]*odd, even + factor[n//2:]*odd])

x = np.array([1, 0, 1, 0,1,0,1,0], dtype = complex)
X = fft(x)
print(X)

# Check value using numpy.fft
X_numpy = np.fft.fft(x)
print(X_numpy)
```

```python
def fft_step(n): #Find one step of the FFT
    omega = np.exp(-2j*np.pi/n)
    P = np.zeros((n,n),dtype=int)
    row = 0
    for i in range(0,n,2):
        P[row, i] = 1
        row += 1
    for i in range(1,n,2):
        P[row, i] = 1
        row += 1

    F_n2 = fourier_matrix(n//2)
    I_n2 = np.identity(n//2)
    # Diagonal matrix with W^0, W^1, W^2, ... W^(n/2-1)
    D = np.zeros((n//2,n//2),dtype=complex)
    for i in range(n//2):
        D[i,i] = omega**i

    A = np.vstack([np.hstack([I_n2, D]), np.hstack([I_n2, -D])])
    Z = np.zeros((n//2,n//2),dtype=int)
    B = np.vstack([np.hstack([F_n2, Z]), np.hstack([Z, F_n2])])


    print(f'A Matrix: \n{A}')
    print(f'B Matrix: \n{B}')
    print(f'Z Matrix: \n{Z}')
    print(f'Permutation Matrix: \n{P}')
    print(f'Identity Matrix: \n{I_n2}')
    print(f'Diagonal Matrix: \n{D}')

    result = np.dot(np.dot(A, B), P)
    c = np.array([1, 0, 1, 0, 1, 0, 1, 0], dtype = complex)
    
    print(f'FFT Step Matrix: \n{result}')

    print('Transformed Matrix: \n', np.dot(result, c))
n = 8
fft_step(n)
```
