## Write a program to obtain shift matrix P with input.

```python
import numpy as np
def shift_matrix(m,N):
    P = np.zeros((m,m),dtype=int)
    k = 0
    elem = N%m
    while k<m:
        P[k][elem] = 1
        elem=(elem+1)%m
        k+=1
    return P


print(f'Shift Matrix:')
P = shift_matrix(4,1)
print(P)
```

## Circulant Matrix

```python
def circulant_matrix(N,C):
    C_N = np.zeros((N,N),dtype=int)
    for i in range(N):
        C_N[i] = np.roll(C,i)

    return C_N

C_N = circulant_matrix(4,[2,-1,0,-1])
print('Circulant Matrix: \n',C_N)
```

```python
def fourier_matrix(n):
    F = np.zeros((n,n), dtype = complex)
    omega = np.exp((-2j*np.pi)/n)
    for i in range(n):
        for j in range(n):
            if i == 0 or j == 0:
                F[i,j] = 1
            else:
                F[i,j] = omega**(i*j)
    return F

F = fourier_matrix(4)
print('Fourier Matrix:\n',F)

eig_F, eig_val_F = np.linalg.eig(F)
print('\nEigenvalue of F:', eig_val_F)
print('\nEigenvector of F:', eig_F)
```

```python
def fourier_norm(F):
    F_trans = F.transpose()
    norms = np.zeros(len(F),dtype=int)
    for i in range(len(norms)):
        norms[i] = np.linalg.norm(F_trans[i])
    print(f'Norm of Fourier: {norms}')
    print(f'Transpose of Fourier:\n {F_trans}\n')
    return norms,F_trans

def normalised_F(trans_F,norms):
    normalised_F = np.zeros((len(norms),len(norms)),dtype=complex)
    for i in range(len(norms)):
        normalised_F[i] = trans_F[i]/norms[i]
    return normalised_F.transpose()
    

norms,trans_F = fourier_norm(F)
normalised_F = normalised_F(trans_F, norms)

print(f'Normalised F:\n {normalised_F}')
```

## Convolution Rule

```python
import numpy as np
from scipy.fft import fft, ifft

# Define two discrete functions (arrays)
c = np.array([1, 2, 3, 4])
d = np.array([0, 1, 0, 0])

# Compute the convolution of c and d
convolution = np.convolve(c, d)

# Compute the Fourier transform of c and d
F_c = fft(c)
F_d = fft(d)

# Compute the product of Fourier transforms
product_of_ft = F_c * F_d

# Compute the inverse Fourier transform of the product
inverse_ft = ifft(product_of_ft)

print("Convolution:", convolution)
print("Inverse Fourier Transform of the product:", inverse_ft)
```
